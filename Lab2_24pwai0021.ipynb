```python
num1 = float(input("Enter a number"))
num2 = float (input("Enter a number"))
operation = input("enter operation +,-,*,/,**,%: ")
if operation=="+" :
    print (f"{num1+num2}")
elif operation=="-" :
     print (f"{num1-num2}")
elif operation=="*" :
     print (f"{num1*num2}")
elif operation=="/" :
     print (f"{num1/num2:.2f}")
elif operation=="**" :
     print (f"{num1**num2:.2f}")
elif operation=="%" :
    if num2==0:
        print("Cannot divide by zero!")
    else:
        print (f"{num1%num2}")
else:
     print ("Invalid operator")



    
```

```python
snum = int(input("Enter a starting number: "))
enum = int(input("Enter an ending number: "))

while snum <= enum:
    if snum % 7 == 0:
        snum += 1
        continue
    if snum % 2 == 0 and snum % 5 == 0:
        print(f"{snum} is Even & Multiple of 5!")
    elif snum % 2 == 0:
        print(f"{snum} is Even")
    elif snum % 5 == 0:
        print(f"{snum} is Multiple of 5!")
    else:
        print(snum)

    snum += 1
```

```python
password=input("Enter Password: ")
if len(password)<6 :
    print ("Too Short! ")
elif not any(ch.isdigit() for ch in password):
    print("Add at least one number ")
elif not any(ch.isupper() for ch in password):
    print("Add at least one capital letter")
else:
    print("Strong password – good choice!")
    
```
