```python
attempts = 5
while attempts>0:
    pswd = input(f"Enter password")
    count = 0
    if len(pswd)>=8:
        count +=1
    if any(ch.isdigit() for ch in pswd):
        count +=1
    if any(ch.isupper() for ch in pswd):
        count +=1
    if count ==0:
        print(f"Weak password")
    elif count==1:
        print(f"weak") 
    elif count==2:
        print(f"moderate")
    else:
        print(f"you created a Strong password")
        break
    attempts -=1
    print(f"attempts left are {attempts}")
if attempts ==0:
    print(f"Password is not strong enough")
```
