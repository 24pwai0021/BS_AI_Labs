```python
import numpy as np

a1 = np.arange(1,11)
a2 = np.arange(1,10).reshape(3,3)
a3 = np.zeros((4,4))
a4 = np.ones((2,5))

print(a1)
print(a2)
print(a3)
print(a4)
```

```python
arr = np.array([[1,2,3],[4,5,6]])

print(arr.shape)
print(arr.size)
print(arr.dtype)
print(arr.ndim)
```

```python
arr = np.array([10,20,30,40,50,60])

print(arr[0])
print(arr[-1])
print(arr[2])

print(arr[arr>30])
```

```python
arr = np.array([10,20,30,40,50,60])

print(arr[1:5])
print(arr[::-1])
print(arr[::2])

m = np.arange(1,10).reshape(3,3)

print(m[0])
print(m[:,1])
print(m[1:])
```

```python
arr = np.array([1,2,3,4,5])

print(arr*5)
print(arr+10)
print(arr**2)
```

```python
arr = np.arange(1,7).reshape(2,3)

print(arr+10)
print(arr*2)
```

```python
a = np.array([[1,2],[3,4]])
b = np.array([[5,6],[7,8]])

print(a+b)
print(a-b)
```

```python
a = np.array([[1,2],[3,4]])
b = np.array([[5,6],[7,8]])

print(np.dot(a,b))
print(a.T)
```

```python
r1 = np.random.randint(1,101,(4,4))
r2 = np.random.rand(4,4)

print(r1)
print(r2)

print(r1.max())
print(r1.min())
print(r1.mean())
```

```python
marks = np.array([45,67,89,90,55,78,82,49,60,95])

print(marks.mean())
print(marks.max())
print(marks.min())

print(len(marks[marks>80]))
print(marks[marks<50])
```

```python
img = np.random.randint(0,256,(5,5))

print(img)
print(img.max())
print(img.min())
print(img.shape)
```

```python
data = np.array([10,20,30,40,50])

norm = (data - data.min()) / (data.max() - data.min())

print(norm)
```

```python
data = np.array([
    [22,25000,0],
    [25,30000,1],
    [30,40000,1],
    [35,50000,1],
    [28,27000,0]
])

X = data[:,:2]
y = data[:,2]

print(X)
print(y)

print(X[:,1].mean())
print(X[:,0].max())
print(data.shape)
```
