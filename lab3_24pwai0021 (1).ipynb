```python
marks1 = int(input("Enter number of english: "))
marks2 = int(input("Enter number of urdu: "))
marks3 = int(input("Enter number of maths: "))
marks = [marks1, marks2, marks3]
def total(m):
    return sum(m)
def average(m):
    return sum(m) / len(m)
highest = (lambda x: max(x))(marks)
print("Total:", total(marks))
print("Average:", round(average(marks), 2))
print("Highest:", highest)
```
